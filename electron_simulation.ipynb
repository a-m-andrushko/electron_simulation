```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# SIMULATION CHECK AND OSCILLATION FIELD MONITORING

# LOAD DATA
quantities = np.loadtxt('quantities_9.txt')
density_1 = np.loadtxt('density_9.txt')
#density_2 = np.loadtxt('density_39,5_2.txt')
#density_3 = np.loadtxt('density_39,5_3.txt')

# EXTRACT QUANTITIES
step = quantities[:, 0] # TIME
N = quantities[:, 1] # NORM
x = quantities[:, 2] # <POSITION>
E = quantities[:, 3] # <ENERGY>

k_1 = density_1[:, 0] # ITERATOR
rho_1 = density_1[:, 1] # DENSITY

#k_2 = density_2[:, 0] # ITERATOR
#rho_2 = density_2[:, 1] # DENSITY

#k_3 = density_3[:, 0] # ITERATOR
#rho_3 = density_3[:, 1] # DENSITY

# PLOT
fig, axs = plt.subplots(2, 2, figsize=(7.5, 7.5))

axs[0, 0].plot(step, N)
axs[0, 0].set_title('N')
axs[0, 0].set_xlim(0.0, 0.05)
axs[0, 0].set_xlabel('STEP')
axs[0, 0].set_ylabel('N')

axs[0, 1].plot(step, x)
axs[0, 1].set_title('<x>')
axs[0, 1].set_xlim(0.0, 0.05)
axs[0, 1].set_xlabel('STEP')
axs[0, 1].set_ylabel('<x>')

axs[1, 0].plot(step, E)
axs[1, 0].set_title('E')
axs[1, 0].set_xlim(0.0, 0.05)
axs[1, 0].set_xlabel('STEP')
axs[1, 0].set_ylabel('E')

axs[1, 1].plot(k_1, rho_1, label = 'Faza 1')
#axs[1, 1].plot(k_2, rho_2, label = 'Faza 2')
#axs[1, 1].plot(k_3, rho_3, label = 'Faza 3')
axs[1, 1].set_title('ρ')
axs[1, 1].set_xlabel('k')
axs[1, 1].set_ylabel('ρ')

#plt.legend(loc = 'upper right')
plt.tight_layout()
plt.show()
```
